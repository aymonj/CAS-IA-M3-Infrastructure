```sql
select
        c.FirstName       as Prenom,
        c.LastName        as Nom,
        sum(soh.TotalDue) as total
from
        saleslt.customer as c
inner join
        saleslt.salesorderheader as soh
	on
        c.customerid = soh.customerid
group by
        c.FirstName,
        c.LastName
order by
        total desc
limit 3
```
