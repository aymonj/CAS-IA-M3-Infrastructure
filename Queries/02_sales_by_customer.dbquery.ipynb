```sql
select
        c.LastName              nom,
        c.FirstName             prenom,
        count(soh.SalesOrderID) as nb_orders
from
        saleslt.customer c
left join
        saleslt.SalesOrderHeader soh
on
        c.CustomerID = soh.CustomerID
group by
        c.LastName,
        c.FirstName
order by
        nb_orders desc
```
