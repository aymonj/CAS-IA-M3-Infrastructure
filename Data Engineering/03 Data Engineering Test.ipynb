```python
# Run 02 Data Engineering Load Silver
import re

with open("./02 Data Engineering Load Silver.sql", "r") as f:
    sql_script = f.read()

# Split by semicolons, but keep multi-line statements together
statements = [stmt.strip() for stmt in sql_script.split(';') if stmt.strip()]

print(f"Executing {len(statements)} SQL statements...\n")

for i, stmt in enumerate(statements, 1):
    if stmt:
        try:
            print(f"[{i}/{len(statements)}] Executing...")
            result = spark.sql(stmt)
            # Only show result if it returns data
            if result is not None:
                result.show(5, truncate=False)
        except Exception as e:
            print(f"Error in statement {i}: {e}")
            print(f"Statement: {stmt[:100]}...")
            raise

print("\n✓ All statements executed successfully!")
```
